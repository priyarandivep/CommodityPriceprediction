```python
import requests
from bs4 import BeautifulSoup

ph_name = []
ph_price = []
ph_descriptions = []

for i in range(1,50):
    url = "https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DAPPLE&p%5B%5D=facets.brand%255B%255D%3Drealme&p%5B%5D=facets.brand%255B%255D%3DPOCO&p%5B%5D=facets.brand%255B%255D%3DOPPO&p%5B%5D=facets.brand%255B%255D%3Dvivo&p%5B%5D=facets.brand%255B%255D%3DREDMI&p%5B%5D=facets.brand%255B%255D%3DMOTOROLA&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.type%255B%255D%3DSmartphones&page=" + str(i)
    req = requests.get(url)
    content = BeautifulSoup(req.content, "html.parser")

    names = content.find_all("div", {"class": "_4rR01T"})
    for name in names:
        ph_name.append(name.text)
        
    price_divs = content.find_all("div", {"class": "_30jeq3 _1_WHN1"})
    for price_div in price_divs:
        price_text = price_div.get_text().replace("₹", "").replace(",", "").strip()
        if price_text:
            ph_price.append(int(price_text))
        else:
            ph_price.append(0)
    descriptions = content.find_all("ul", {"class": "_1xgFaf"})  # Update selector if needed
    for description in descriptions:
        li_elements = description.find_all("li")
        if len(li_elements) >= 4:
            ph_descriptions.append([li.text.strip() for li in li_elements[:4]])
        else:
            ph_descriptions.append([])  # Add an empty list if less than 4 "li" elements are found

    while len(ph_price) < len(ph_name):
        ph_price.append(0)

# Verify that the lengths of ph_name and ph_price lists match
print("Length of ph_name:", len(ph_name))
print("Length of ph_price:", len(ph_price))
print("Length of ph_description:", len(ph_descriptions))
```

```python
data = {
    "Name": ph_name,
    "Price": ph_price,
    "ph_description": ph_descriptions
}
```

```python
# Extract RAM information from p1 using regular expressions
rom_pattern = r'(\d+\s?GB\s?ROM)'
df['Rom'] = df['p1'].apply(lambda x: re.findall(rom_pattern, x)[0] if re.findall(rom_pattern, x) else None)
```

```python
# Extract display information from p2 using regular expressions
display_pattern = r'(\d+(?:\.\d+)?\s?cm)'

df['display'] = df['p2'].apply(lambda x: re.findall(display_pattern, x)[0] if re.findall(display_pattern, x) else None)
```

```python
def has_hd(description):
    if "HD" in description:
        return "Yes"
    else:
        return "No"
```

```python

def extract_front_camera_resolution(description):
    pattern = r'(\d+)MP Front Camera'
    match = re.search(pattern, description)
    if match:
        return int(match.group(1))
    else:
        return None

df["Front_Camera_Resolution_MP"] = df["p3"].apply(extract_front_camera_resolution)
```
